## 2 Forest Fire Sampling II

#### Building the model

Models consist of __agents__ and the __environment__
In the forest fire model, the agent is a tree and it's state can be
- fine(unburned)
- on fire
- burnt


The environment is the graph of pinterest users.


BFS for graph to spread fire


```python
import numpy as np
from collections import deque
import pymongo
from pymongo import MongoClient
print( 'Mongo version', pymongo.__version__)
```

```python
#main db
config = {
  'host': '127.0.0.1',
  'username': 'mongo_user',
  'password': 'password',
  'authSource': 'pin'
}

client = MongoClient(**config)

db = client.pin
collection1=db.users
collection2=db.sampled_users
collection3=db.more_sampled
```

```python
### Forest Fire Model
```

```python
explored = []
sampled=[]
```

```python
def burn_or_not(p = 0.9):
    return (np.random.random() <=p)

def select_node():
    node = str(np.random.randint(1,50000000))
    if node not in sampled:
        return node

def spread_the_fire_bfs(obj):
    sampled.append(obj['user_id'])
    burn = burn_or_not()
    print(obj)
    # keep track of explored nodes
    # keep track of all the paths to be checked
    if burn:
        print("burn", flush=True)
        queue = [obj]
        #print(queue)
        while queue:
            # pop first node from queue
            node = queue.pop(0)
            print(node)
            if node['user_id'] not in explored:
                #add node to list of explored nodes
                explored.append(node['user_id'])
                print(explored)
                #insert obj to collection2
                collection3.insert_one(node)
                newobj = collection1.find_one({'user_id':node['user_id']})
                burn = burn_or_not()
                if burn:
                    queue.extend(newobj['followers'])
                    
            if collection3.find().count() > 1000:
                return
       
        
    else:
        print("no burn", flush=True)
        start_the_fire()
        
def start_the_fire():
    try:
        user_id = select_node()
        obj = collection1.find_one({'user_id':user_id})
        spread_the_fire_bfs(obj)
    except:
        pass
        
    return 
```

```python
try_burning= True
while try_burning:
    start_the_fire()
    if collection3.find().count() == 1000:
        try_burning = False
    
```
